```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('HR_EA.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})

```

```python
x=df[['Age','OverTime','JobSatisfaction','WorkLifeBalance','MonthlyIncome','JobLevel','YearsAtCompany','TotalWorkingYears','DistanceFromHome']]
y=df['Attrition']
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

scaler=StandardScaler()
scaler.fit(x_train)
x_trs=scaler.transform(x_train)
x_tes = scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_trs,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_tes,columns=x_test.columns)
```
